```python
import jieba.analyse
import json
jieba.load_userdict("mydict.txt")

num = 181

with open("printed_news_30", "r", encoding="utf-8") as file:
    data = json.load(file)
file.close()
article = []
for c in data:
    article.append(c["content"])
split = " ".join(jieba.cut(article[num]))
print(split)
```
